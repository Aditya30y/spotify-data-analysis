### code to find least popular song. 
### To find the most popular song, ascending = false to be used so that it gives result in descending manner


```python

popular_df = df_tracks.sort_values('popularity',ascending = True).head(10)
popular_df
```

```python
df_tracks.describe().transpose()
```

#### to find most popular song with certain criteria

```python

most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending= False)
most_popular[:10]
```

#### code to change datetime format to yyyy-mm-dd

```python


df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

#### check artist at 50th number

```python


df_tracks[['artists']].iloc[50]
```

#### changing millisecond to second to give understandable and real life data

```python


df_tracks['duration_second']=df_tracks['duration_ms'].apply(lambda x : round(x/1000))
df_tracks.drop('duration_ms', inplace =True, axis =1)
df_tracks.head()
```

### to find correlation among different parameters 
#### Pearson method used assigns a value to 1 to -1 ,  where 0 is no correlation, 1 is total positive correlation, and − 1 is total negative correlation 

```python
corr_df=df_tracks.drop(['key','explicit','mode'],axis =1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1, center=0, cmap= "inferno", linewidths =1, linecolor = "Black")
heatmap.set_title(" correlation heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
# to get a Particular sample size of dataset 


sample_df = df_tracks.sample(int(.006*len(df_tracks)))
print(len(sample_df))
```

 ### to find correlation between loudness and energy using various attributes


```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y ='loudness', x='energy',color ='c').set(title= "loudness vs energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y ='popularity', x='danceability',color ='y').set(title= "popularity vs danceability correlation")
```

#### creating a new column dates 

```python


df_tracks['dates'] =df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
df_tracks.head()
```

#### visualising no. of songs per year using a histogram

```python
sns.displot(years, discrete = True, aspect =2,height =5, kind="hist").set(title = "number of songs per year")
```

### visulaisng song length per year using barplot

```python
total_dr=df_tracks.duration_second
figdims=(18,7)
fig , ax =plt.subplots(figsize =figdims)
fig = sns.barplot(x =years,y=total_dr, ax=ax ,errwidth = False).set(title = " song length per year")
plt.xticks(rotation = 90)
```

### visulaisng song length per year using lineplot

```python
total_dr=df_tracks.duration_second
sns.set_style(style = 'whitegrid')
figdims=(10,6)
fig , ax =plt.subplots(figsize =figdims)
fig = sns.lineplot(x =years,y=total_dr, ax=ax ).set(title = " song length per year")
plt.xticks(rotation = 60)
```

#### importing second dataset

### calculating which genre has the longest duration of song. 

```python
plt.title("duration of songs acc to genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y= 'genre',x='duration_ms',data=df_genre)
plt.xlabel("duration in ms")
plt.ylabel("genres")
```
